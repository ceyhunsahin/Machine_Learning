```python
import pandas as pd       
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('Hitters.csv')
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
df
```

```python
df.dropna(inplace = True)
df
```

```python
df = df._get_numeric_data() # PCA'nin temeli budur
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df = StandardScaler().fit_transform(df)
```

```python
df[0:5,0:5]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca_fit = pca.fit_transform(df)
```

```python
component_df = pd.DataFrame(data = pca_fit, columns = ['first_component', 'second_component'])
component_df
```

```python
sns.pairplot(component_df)
```

```python
# EXPLAINED VARIANCE RATIO
pca.explained_variance_ratio_ # 0.45 1. nin varyansi, yuzde 45 dogru temsil var, 
#digerinde 24 temsil toplamda yuzde 70 temsil soz konusu
```

```python
# OPTIMUM k Sayisini bulma
pca = PCA().fit(df)
X = range(1,18) # graph 1 den baslasin diye
plt.plot(X,np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('component count')
plt.ylabel('Variance Ratio');
```
